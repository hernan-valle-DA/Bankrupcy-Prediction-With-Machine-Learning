## Búsqueda de Outliers

```python
variables = data.drop('Bankrupt?',axis=1).columns

def es_outlier(df, columna, umbral=1.5):
  q1 = df[columna].quantile(0.25)
  q3 = df[columna].quantile(0.75)
  iqr = q3 - q1
  return (df[columna]<q1-umbral*iqr)|(df[columna]>q3+umbral*iqr) 

contador = 0

print('Cantidad de outliers por columna:')
for i in variables:
  x = data.loc[es_outlier(data, i)].shape[0]
  contador += 1
  print(f'{contador}- {i}: {x}')
```

#### En este punto se aplicarán dos estrategias para abordar el desbalanceo: SMOTE y Undersampling.

### Smote

#### La técnica SMOTE (Synthetic Minority OverSampling Technique) consiste en sintetizar nuevos ejemplos de la clase minoritaria. SMOTE selecciona ejemplos cercanos en el espacio de características, traza una línea entre ellos y crea un nuevo ejemplo en un punto de esa línea. Esto se hace eligiendo un ejemplo aleatorio de la clase minoritaria y encontrando sus k vecinos más cercanos, luego se selecciona un vecino al azar y se crea un ejemplo sintético entre los dos.

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy=.5, random_state=30)

X_train_resample, y_train_resample = smote.fit_resample(X_train, y_train)
```

#### Se verifica como queda la proporción de la variable objetivo

```python
y_train_resample.value_counts(normalize=True)*100
```

```python
balance_over = y_train_resample.value_counts()
plt.figure(figsize=(10,5))
sns.barplot(balance_over)
plt.title('Conteo de datos de la variable objetivo "Bankrupt?"')
plt.show()
```

### Random Undersampling

#### El random undersampling consiste en seleccionar aleatoriamente ejemplos de la clase mayoritaria para eliminarlos del conjunto de datos de entrenamiento, reduciendo así el número de ejemplos en la clase mayoritaria hasta lograr la distribución deseada.

```python
from imblearn.under_sampling import RandomUnderSampler
undersample = RandomUnderSampler(sampling_strategy=1)

X_train_under, y_train_under = undersample.fit_resample(X_train_resample, y_train_resample)
```

#### Se verifica como queda la proporción de la variable objetivo

## Imputación de Outliers

#### Para imputar los outlier se utiliza el método de Flooring and Capping: los valores menores serán reemplazados por primer cuartil menos 1.5 veces el rango intercuartil y los valores mayores serán reemplazados por el tercer cuartil más 1.5 veces el rango intercuartil.

```python
def imputador_outlier(data,variables):
    
    outliers = data.copy()
    
    for i in variables:        
        Q1 = outliers[i].quantile(0.25)
        Q3 = outliers[i].quantile(0.75)
        IQR = Q3 - Q1
        valor_menor = (Q1 - 1.5 * IQR)
        valor_mayor = (Q3 + 1.5 * IQR)
        
        outliers.loc[outliers[i] < valor_menor,i] = valor_menor
        outliers.loc[outliers[i] > valor_mayor,i] = valor_mayor
        
    return outliers 

nuevos_datos = X_train_selected_2.select_dtypes(include='float')
variables2 = nuevos_datos.columns

X_train_imp = imputador_outlier(nuevos_datos,variables2)  
```

```python
data_ent = X_train_selected_2.select_dtypes(include='int')

X_train_mod = pd.concat([X_train_imp,data_ent],axis=1)

X_train_mod.head()
```
